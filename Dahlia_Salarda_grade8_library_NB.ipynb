```python
import requests
import json
"""
The 'students' variable is already loaded from the URL in a previous cell.
Therefore, we can directly use the 'students' variable.
"""
data = books
print("JSON file loaded")
ref = db.reference("books")
for book in data["books"]:ref.child(str(book["id"])).set(book)
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
total_copies = 0
for book_item in books_data.values():
  total_copies += book_item['details']['copies']

print(f"Total copies of all books: {total_copies}")
```

```python
print("Book Titles and Availability:")
for book_id, book_info in books_data.items():
    title = book_info['title']
    availability = book_info['details']['available']
    print(f"Title: {title}, Available Copies: {availability}")
```
